```python
FEATURES = ['MIN','MAX','MEAN','RMS','VAR','STD','POWER','PEAK','P2P','CREST FACTOR','SKEW','KURTOSIS']

def features_extraction(x): 
    
    
    Min=[];Max=[];Mean=[];Rms=[];Var=[];Std=[];Power=[];Peak=[];Skew=[];Kurtosis=[];P2p=[];CrestFactor=[];
    FormFactor=[]; PulseIndicator=[];
    
    X = x.values
    X_scaled = X
    ## TIME DOMAIN ##

    Min.append(np.min(X_scaled))
    Max.append(np.max(X_scaled))
    Mean.append(np.mean(X_scaled))
    Rms.append(np.sqrt(np.mean(X_scaled**2)))
    Var.append(np.var(X_scaled))
    Std.append(np.std(X_scaled))
    Power.append(np.mean(X_scaled**2))
    Peak.append(np.max(np.abs(X_scaled)))
    P2p.append(np.ptp(X_scaled))
    CrestFactor.append(np.max(np.abs(X))/np.sqrt(np.mean(X_scaled**2)))
    Skew.append(stats.skew(X_scaled))
    Kurtosis.append(stats.kurtosis(X_scaled))
    FormFactor.append(np.sqrt(np.mean(X_scaled**2))/np.mean(X_scaled))
    PulseIndicator.append(np.max(np.abs(X_scaled))/np.mean(X_scaled))
   
    #Create dataframe from features
    x_features = pd.DataFrame(index = [FEATURES], 
                               data = [Min,Max,Mean,Rms,Var,Std,Power,Peak,P2p,CrestFactor,Skew,Kurtosis])
    return x_features
```

```python
X = data1.iloc[:,0:-1]
y = data1.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.2, random_state=0) 
```

```python
models = []
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('DTC', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
# evaluate each model in turn
results = []
names = []
for name, model in models:
	kfold = StratifiedKFold(n_splits=10, random_state=None, shuffle=True)
	cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring='accuracy')
	results.append(cv_results)
	names.append(name)
	print('%s: %f (%f)' % (name, (cv_results.mean())*100, cv_results.std()))
```
